# Hidden Markov Model Implementation
In this notebook, we discretize the features using K-Means and implement the HMM (Viterbi & Baum-Welch) from scratch.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

# Load features
df = pd.read_csv('processed_features.csv')

# Discretize using K-Means
# We will use K=20 observation symbols
n_symbols = 20
feature_cols = [c for c in df.columns if c not in ['activity', 'session', 'window_id']]

kmeans = KMeans(n_clusters=n_symbols, random_state=42)
df['symbol'] = kmeans.fit_predict(df[feature_cols])

# Map activities to state integers
activities = ['standing', 'walking', 'jumping', 'still']
state_map = {act: i for i, act in enumerate(activities)}
df['state'] = df['activity'].map(state_map)

# Split into train/test sessions
sessions = df['session'].unique()
train_sessions, test_sessions = train_test_split(sessions, test_size=0.2, random_state=42)

train_df = df[df['session'].isin(train_sessions)]
test_df = df[df['session'].isin(test_sessions)]

train_df.to_csv('train_data.csv', index=False)
test_df.to_csv('test_data.csv', index=False)

print(f"Train size: {len(train_df)} windows, Test size: {len(test_df)} windows")
```

```python
# HMM Implementation from scratch
class HMM:
    def __init__(self, n_states, n_symbols):
        self.n_states = n_states
        self.n_symbols = n_symbols
        self.pi = np.ones(n_states) / n_states
        self.A = np.ones((n_states, n_states)) / n_states
        self.B = np.ones((n_states, n_symbols)) / n_symbols
        
    def _forward(self, obs):
        T = len(obs)
        alpha = np.zeros((T, self.n_states))
        c = np.zeros(T) # Scaling factors to prevent underflow
        
        alpha[0] = self.pi * self.B[:, obs[0]]
        c[0] = 1.0 / (np.sum(alpha[0]) + 1e-10)
        alpha[0] *= c[0]
        
        for t in range(1, T):
            alpha[t] = np.dot(alpha[t-1], self.A) * self.B[:, obs[t]]
            c[t] = 1.0 / (np.sum(alpha[t]) + 1e-10)
            alpha[t] *= c[t]
            
        return alpha, c
        
    def _backward(self, obs, c):
        T = len(obs)
        beta = np.zeros((T, self.n_states))
        
        beta[T-1] = 1.0 * c[T-1]
        
        for t in range(T-2, -1, -1):
            beta[t] = np.dot(self.A, self.B[:, obs[t+1]] * beta[t+1])
            beta[t] *= c[t]
            
        return beta
        
    def fit(self, observations, max_iter=100, tol=1e-4):
        log_likelihoods = []
        
        for it in range(max_iter):
            pi_num = np.zeros(self.n_states)
            A_num = np.zeros((self.n_states, self.n_states))
            A_den = np.zeros(self.n_states)
            B_num = np.zeros((self.n_states, self.n_symbols))
            B_den = np.zeros(self.n_states)
            
            total_log_likelihood = 0
            
            for obs in observations:
                T = len(obs)
                alpha, c = self._forward(obs)
                beta = self._backward(obs, c)
                
                total_log_likelihood += -np.sum(np.log(c + 1e-10))
                
                gamma = alpha * beta / (c.reshape(-1, 1) + 1e-10)
                
                pi_num += gamma[0]
                
                for t in range(T - 1):
                    xi = (alpha[t].reshape(-1, 1) * self.A * self.B[:, obs[t+1]] * beta[t+1])
                    A_num += xi
                    A_den += gamma[t]
                    
                for t in range(T):
                    B_num[:, obs[t]] += gamma[t]
                    B_den += gamma[t]
                    
            self.pi = pi_num / np.sum(pi_num)
            for i in range(self.n_states):
                self.A[i] = A_num[i] / (A_den[i] + 1e-10)
                self.B[i] = B_num[i] / (B_den[i] + 1e-10)
                
            log_likelihoods.append(total_log_likelihood)
            print(f"Iteration {it+1}, Log-Likelihood: {total_log_likelihood:.4f}")
            
            if it > 0 and abs(log_likelihoods[-1] - log_likelihoods[-2]) < tol:
                print("Converged!")
                break
                
    def viterbi(self, obs):
        T = len(obs)
        delta = np.zeros((T, self.n_states))
        psi = np.zeros((T, self.n_states), dtype=int)
        
        delta[0] = np.log(self.pi + 1e-10) + np.log(self.B[:, obs[0]] + 1e-10)
        
        for t in range(1, T):
            for j in range(self.n_states):
                prob = delta[t-1] + np.log(self.A[:, j] + 1e-10)
                psi[t, j] = np.argmax(prob)
                delta[t, j] = np.max(prob) + np.log(self.B[j, obs[t]] + 1e-10)
                
        states = np.zeros(T, dtype=int)
        states[T-1] = np.argmax(delta[T-1])
        
        for t in range(T-2, -1, -1):
            states[t] = psi[t+1, states[t+1]]
            
        return states
```

```python
# Supervised Initialization
# We use the training labels to initialize the A and B matrices to ensure State 0 maps to Activity 0, etc.
hmm = HMM(n_states=4, n_symbols=20)

# Initialize B by counting symbols per activity
B_init = np.ones((4, 20)) # Additive smoothing
for state in range(4):
    state_data = train_df[train_df['state'] == state]
    counts = state_data['symbol'].value_counts()
    for sym, count in counts.items():
        B_init[state, sym] += count
    hmm.B[state] = B_init[state] / np.sum(B_init[state])
    
# Initialize A with strong diagonal (e.g., highly likely to stay in the same activity)
A_init = np.ones((4, 4)) * 0.05
np.fill_diagonal(A_init, 0.85)
hmm.A = A_init

# Initialize Pi uniformly
hmm.pi = np.array([0.25, 0.25, 0.25, 0.25])

# Prepare observation sequences for training
# We concatenate sessions to allow the HMM to learn transitions
train_sequences = []
for session, group in train_df.groupby('session'):
    train_sequences.append(group['symbol'].values)
    
long_sequences = []
chunk_size = max(1, len(train_sequences) // 5)
for i in range(5):
    combined = np.concatenate(train_sequences[i*chunk_size:(i+1)*chunk_size])
    long_sequences.append(combined)

print("Training HMM with Baum-Welch...")
hmm.fit(long_sequences, max_iter=50, tol=1e-3)

# Save parameters
np.savez('hmm_params.npz', pi=hmm.pi, A=hmm.A, B=hmm.B)
print("HMM trained and parameters saved to hmm_params.npz")
```
